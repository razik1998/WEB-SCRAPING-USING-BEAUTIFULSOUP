```python
# Importing Required Labraries 

from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
# Send Get Request to the Web Page Server

page = requests.get ('https://en.wikipedia.org/wiki/Main_Page')
page
```

```python
# Only Need Particular Content From Web-Site

soup = BeautifulSoup (page.content)
soup
```

```python
#  Scraping All Header Tags In One_Go
Header_tag = []  
for i in soup.find_all ('span',class_="mw-headline"):
    Header_tag.append (i.text)
    
```

```python
#  Displaying All Header_Tags
Header_tag
```

```python
# Making Data Frame & Displaying In Table Form

df = pd.DataFrame ({'All_Headlines':Header_tag})
df
```
